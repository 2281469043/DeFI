## **Step 1: Building an Existing RNN Model**

Wil use tensorflow and LSTM first.

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Data Preprocessing
# Encode categorical variables
label_encoder = LabelEncoder()
borrow['user'] = label_encoder.fit_transform(borrow['user'])
```

```python
# Scale numerical variables
scaler = StandardScaler()
numerical_cols = ['amount', 'amountUSD', 'amountETH']
borrow[numerical_cols] = scaler.fit_transform(borrow[numerical_cols])
```

```python
# Organize data into sequences
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length + 1):
        sequence = data[i:i + seq_length]
        sequences.append(sequence)
    return np.array(sequences)
```

```python
# Define sequence length
seq_length = 10
# Create sequences for training
X = create_sequences(
    borrow[['user', 'amount', 'amountUSD', 'amountETH']].values, seq_length)
y = borrow['amount'].values[seq_length - 1:]
```

```python
# Define Training and Validation Sets
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, shuffle=False)
```

```python
# Build the RNN Model
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(units=64, input_shape=(
        X_train.shape[1], X_train.shape[2])),
    tf.keras.layers.Dense(units=1)
])
```

```python
# Compile the Model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
# Train the Model
history = model.fit(X_train, y_train, epochs=5,
                    batch_size=64, validation_data=(X_val, y_val))
```

```python
# Evaluate the Model
mse, mae = model.evaluate(X_val, y_val)
print(f'Validation MSE: {mse}, Validation MAE: {mae}')
```

## **Step 2: Update Network with LSTM and GRU**
LSTM Implementation:

```python
# Build the LSTM Model
lstm_model = tf.keras.Sequential([
    tf.keras.layers.LSTM(units=64, input_shape=(
        X_train.shape[1], X_train.shape[2])),
    tf.keras.layers.Dense(units=1)
])

# Compile the LSTM Model
lstm_model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the LSTM Model
lstm_history = lstm_model.fit(
    X_train, y_train, epochs=5, batch_size=64, validation_data=(X_val, y_val))
```

GRU Implementation:

```python
# Build the GRU Model
gru_model = tf.keras.Sequential([
    tf.keras.layers.GRU(units=64, input_shape=(
        X_train.shape[1], X_train.shape[2])),
    tf.keras.layers.Dense(units=1)
])

# Compile the GRU Model
gru_model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the GRU Model
gru_history = gru_model.fit(
    X_train, y_train, epochs=5, batch_size=64, validation_data=(X_val, y_val))
```

**Performance Evaluation and Data Visualization**

Let's evaluate the performance of all three models (Basic RNN, LSTM, GRU) and visualize the loss function during training.

```python
import matplotlib.pyplot as plt

# Plotting Loss Curves
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='RNN Training Loss')
plt.plot(history.history['val_loss'], label='RNN Validation Loss')
plt.plot(lstm_history.history['loss'], label='LSTM Training Loss')
plt.plot(lstm_history.history['val_loss'], label='LSTM Validation Loss')
plt.plot(gru_history.history['loss'], label='GRU Training Loss')
plt.plot(gru_history.history['val_loss'], label='GRU Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
